```python
import statistics

# Function definitions
def calculate_mean(data):
    return statistics.mean(data)

def calculate_median(data):
    return statistics.median(data)

def calculate_mode(data):
    try:
        return statistics.mode(data)
    except statistics.StatisticsError:
        return "No unique mode found"

def calculate_geometric_mean(data):
    return statistics.geometric_mean(data)

def calculate_harmonic_mean(data):
    return statistics.harmonic_mean(data)


# Main Program Loop
while True:
    print("\nCentral Tendency Calculator")
    print("1. Mean")
    print("2. Median")
    print("3. Mode")
    print("4. Geometric Mean")
    print("5. Harmonic Mean")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == "6":
        print("Exiting program... Goodbye! üëã")
        break

    # User enters dataset
    try:
        data = list(map(float, input("Enter numbers separated by space: ").split()))
    except ValueError:
        print("‚ö†Ô∏è Invalid input! Please enter numeric values only.")
        continue

    if choice == "1":
        print("Mean:", calculate_mean(data))
    elif choice == "2":
        print("Median:", calculate_median(data))
    elif choice == "3":
        print("Mode:", calculate_mode(data))
    elif choice == "4":
        print("Geometric Mean:", calculate_geometric_mean(data))
    elif choice == "5":
        print("Harmonic Mean:", calculate_harmonic_mean(data))
    else:
        print("‚ö†Ô∏è Invalid Choice! Please select between 1‚Äì6.")
```
